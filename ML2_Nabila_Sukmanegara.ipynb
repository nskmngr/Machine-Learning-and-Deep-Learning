### Lakukan segmentasi data dengan metode Clustering, metode yang digunakan bebas. Hasil akhir akan dievaluasi menggunakan fungsi silhouette score seperti contoh di bawah ini

```python
#Lakukan Proses Pencarian CLuster terbaik untuk data diatas

#Import Library
from sklearn.cluster import KMeans
from sklearn import metrics
from warnings import filterwarnings as fw; fw('ignore')

#START CODE

#Inertia adalah Pembuatan sebuah list kosong dengan nama
#Inertia yang akan digunakan untuk menyimpan nilai inersia untuk setiap nilai k (jumlah cluster) yang dicoba.
inertia = []

#mencoba berbagai nilai k mulai dari 10 hingga 24
#setelah dilakukan percobaan, k terbaik adalah 15 jadi ambil di tengah tengah
for k in range(10, 25):
    #Melakukan clustering pada data
    model = KMeans(n_clusters=k, random_state=42)
    #Melakukan clustering pada data meggunakan model K-Means yang telah dibuat
    #Setelah loop selesai dieksekusi, list inertia akan berisi nilai inersia untuk berbagai nilai k yang telah dicoba.
    model.fit(df)
    inertia.append(model.inertia_)
inertia
#END CODE
```

Jumlah cluter terbaik ada 15.


Kode ini adalah langkah awal dalam mencari jumlah cluster terbaik (k) dengan menggunakan metode Elbow atau metode lainnya. Hasil nilai inersia ini dapat digunakan untuk menganalisis dan memilih jumlah cluster yang optimal untuk data yang sedang diolah.

```python
inertia = pd.DataFrame({'Inertia': inertia}, index=[i for i in range(10, 25)])
print(inertia)
```

```python
# Plot nilai inersia

#Ukuran grafik
plt.figure(figsize=(8, 4))
#Membuat plot garis pada grafik
inertia.plot(kind='line', marker='o', linestyle='--')
plt.ylabel('Inertia')
plt.xlabel('Jumlah Cluster (k)')
plt.title('Metode Elbow untuk Menentukan Jumlah Cluster Terbaik')
plt.grid(True)
plt.show()
```

- Sumbu x menunjukkan jumlah cluster (k) yang dicoba dalam rentang 10 hingga 25. Setiap titik pada sumbu x mewakili jumlah cluster yang sedang diuji.
- Sumbu y menunjukkan nilai inersia yang dihasilkan oleh algoritma K-Means untuk setiap nilai k. Nilai inersia adalah jumlah jarak kuadrat antara data point dengan pusat cluster terdekat.
- Titik-titik pada grafik ditandai dengan marker 'o' (lingkaran) dan garis putus-putus ('--') yang menghubungkan titik-titik tersebut. Marker 'o' memungkinkan kita untuk melihat dengan jelas di mana titik data terletak pada grafik.

```python
#Import library
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd

# Jumlah cluster terbaik yang sudah ditentukan
best_K = 15
# Inisialisasi model K-Means dengan jumlah cluster terbaik
kmeans = KMeans(n_clusters=best_K, random_state=0)
# Melakukan clustering pada data
label = kmeans.fit_predict(df)
# Menghitung Silhouette Score
score = silhouette_score(df, label)
# Menampilkan hasil Silhouette Score
print("Silhouette Score =", score)
```

Jika Silhouette Score adalah nilai positif, seperti dalam contoh (0.711278614093076), itu adalah indikasi bahwa clustering memiliki kualitas yang baik.

### Proses Visualisasi hasil CLuster

```python
#Proses Visualisasi hasil cluster

#Pertama, masukkan dahulu label hasil clustering ke dalam data
result = df.copy()

#Menambahkan kolom cluster
result['cluster'] = label
#Menampilkan 5 baris pertama dari data
result.head()
#END CODE
```
