# Preparation

Preparation biasa dilakukan untuk mempersiapkan data sebelum masuk dalam tahap pemodelan. <br>

## Import Library

```python
import pandas as pd
import numpy as np
```

## Input Dataset

```python
#Membaca data dan memasukkannya ke dalam bentuk Pandas Dataframe
df = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/SC_HW1_bank_data.csv')
```

```python
#Jalankan code untuk mengecek nama kolom yang tersedia
df.columns
```

```python
#Menampilkan data
df.head(10)
```

```python
#menampilkan ukuran data
df.shape
```

```python
#Hilangkan kolom yang dirasa tidak relevan dengan model (contoh: ID). None dapat diisi dengan nama-nama kolom yang akan digunakan.
#Contoh df = df[['X1','X2', 'Y']].copy()

#START CODE
# Hilangkan kolom 'RowNumber' dan pilih kolom-kolom yang relevan
df = df[['Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'Exited']].copy()
df.head()
```

## Preprocessing

```python
#Lakukan One-Hot Encoder untuk data categorical, dengan fungsi pandas get_dummies
# Menggunakan get_dummies untuk melakukan One-Hot Encoding pada kolom-kolom kategori
df_encoded = pd.get_dummies(df, columns=['Geography', 'Gender'])

# Menampilkan 5 baris pertama data yang telah diencode
df_encoded.head()
```

```python
#Pisahkan variabel fitur (X) dan variabel target (Y)

#X berisi semua kolom kecuali "Exited"
X = df_encoded.drop(columns=['Exited'])
#Y berisi hanya kolom "Exited"
y = df_encoded['Exited']
```

```python
#Menampilkan beberapa contoh data fitur (X)
X.head()
```

```python
#Menampilkan conton data target (y)
y.head()
```

```python
#Lakukan Scaler dan/atau Noermalisasi jika diperlukan
from sklearn.preprocessing import MinMaxScaler

# Inisialisasi MinMaxScaler
scaler = MinMaxScaler()

# Melakukan normalisasi pada fitur (X_encoded)
X_transform = scaler.fit_transform(X)

# X_transform adalah hasil fitur yang telah dinormalisasi
```

## Train-Test Split

```python
#Split menjadi train dan test dengan test_size 25%
#Tidak perlu mengubah code ini

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_transform,y,test_size = 0.25,random_state = 123)
```

# Modeling

## Model1

### Soal :
Jelaskan secara Singkat Model pertama yang digunakan!

```python
# Import library scikit-learn
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Inisialisasi model Decision Tree
model1 = DecisionTreeClassifier(random_state=42)  # Tambahkan random_state untuk hasil yang dapat direproduksi

# Definisikan daftar hyperparameter yang akan diuji
params = {
    "criterion": ["gini", "entropy"],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

# Buat objek GridSearchCV untuk mencari parameter terbaik
grid = GridSearchCV(
    estimator=model1,
    param_grid=params,
    scoring='accuracy',
    n_jobs=-1,  # Menggunakan semua core CPU yang tersedia
    cv=10
)

# Melakukan fitting model dengan data training
grid.fit(X_train, y_train)

# Menampilkan parameter terbaik yang ditemukan
best_params = grid.best_params_
print("Parameter terbaik untuk Decision Tree:", best_params)
```

```python
#lakukan evaluasi dengan beberapa maetric di bawah ini
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

**Membaca hasil output :**
- Precision:

Kelas 0 (Tidak Ada): Dari semua prediksi kelas 0, sekitar 88% benar (precision 0.89).
Kelas 1 (Ada): Dari semua prediksi kelas 1, sekitar 69% benar (precision 0.69).

- Recall:

Kelas 0 (Tidak Ada): Dari semua instance yang sebenarnya termasuk dalam kelas 0, model dapat memprediksi sekitar 94% dengan benar (recall 0.94).
Kelas 1 (Ada): Dari semua instance yang sebenarnya termasuk dalam kelas 1, model hanya dapat mendeteksi sekitar 49% (recall 0.49).

- F1-Score:

F1-score mencoba mencari keseimbangan antara precision dan recall. Untuk kelas 0, f1-score adalah 0.91, sedangkan untuk kelas 1, f1-score adalah 0.57.

- Matriks Konfusi:

Model berhasil memprediksi sebagian besar kelas 0 (1871 prediksi benar, 112 prediksi salah).
Model memiliki kesulitan dalam memprediksi kelas 1 (265 prediksi benar, 252 prediksi salah).

- Akurasi:

Akurasi model adalah 0.8492, yang menunjukkan bahwa model ini benar sekitar 84.92% dari waktu.

## Model2
### Soal :
Jelaskan secara Singkat Model ke-2 yang digunakan!

```python
# Import library
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

# Inisialisasi model SVM
model2 = SVC()

# Tentukan hyperparameter yang akan di-tune
param_grid = {
    # Parameter penalti untuk mengendalikan toleransi kesalahan dalam klasifikasi
    'C': [0.1, 1, 10],
    # Jenis kernel dalam SVM
    'kernel': ['linear', 'rbf'],
    # Parameter gamma (kernel rbf)
    'gamma': ['scale', 'auto', 0.1, 1]
}

# Buat objek GridSearchCV
grid = GridSearchCV(
    estimator=model2,
    param_grid=param_grid,
    scoring='accuracy',
    n_jobs=10,
    cv=10
)

# Fit model dan temukan parameter terbaik
grid.fit(X_train, y_train)
best_params = grid.best_params_
print("Parameter terbaik untuk SVM:", best_params)
```

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

**Membaca Hasil output :**
- Precision:

Kelas 0 (Tidak Ada): Dari semua prediksi kelas 0, sekitar 87% benar (precision 0.87).
Kelas 1 (Ada): Dari semua prediksi kelas 1, sekitar 79% benar (precision 0.79).
- Recall:

Kelas 0 (Tidak Ada): Dari semua instance yang sebenarnya termasuk dalam kelas 0, model dapat memprediksi sekitar 97% dengan benar (recall 0.97).
Kelas 1 (Ada): Dari semua instance yang sebenarnya termasuk dalam kelas 1, model hanya dapat mendeteksi sekitar 43% (recall 0.43).
- F1-Score:

F1-score mencoba mencari keseimbangan antara precision dan recall. Untuk kelas 0, f1-score adalah 0.92, sedangkan untuk kelas 1, f1-score adalah 0.55.
- Matriks Konfusi:

Model berhasil memprediksi sebagian besar kelas 0 (1924 prediksi benar, 59 prediksi salah).
Model memiliki kesulitan dalam memprediksi kelas 1 (296 prediksi benar, 221 prediksi salah).
- Akurasi:

Akurasi model adalah 0.858, yang berarti model ini memprediksi kelas dengan benar sekitar 85.8% dari waktu.

## Model3
### Soal :
Jelaskan secara Singkat Model ke-3 yang digunakan!

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# Inisialisasi model KNN
model3 = KNeighborsClassifier()

# Tentukan hyperparameter yang akan di-tune untuk KNN
param_grid = {
    # Jumlah tetangga yang akan dipertimbangkan
    'n_neighbors': [3, 5, 7, 9],
    # Metrik jarak yang akan digunakan. Dua metrik sedang diuji yaitu Euclidean dan Manhattan.
    'metric': ['euclidean', 'manhattan']
}

# Buat objek GridSearchCV untuk KNN
grid = GridSearchCV(
    estimator=model3,
    param_grid=param_grid,
    scoring='accuracy',
    n_jobs=10,
    cv=10
)

# Fit model KNN dan temukan parameter terbaik
grid.fit(X_train, y_train)
best_params = grid.best_params_
print("Parameter terbaik untuk KNN:", best_params)
```

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

- F1-Score:

F1-score mencoba mencari keseimbangan antara precision dan recall. Untuk kelas 0, f1-score adalah 0.90, sedangkan untuk kelas 1, f1-score adalah 0.45.
